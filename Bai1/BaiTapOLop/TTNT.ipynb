```python
"""Bai 11"""
n = int(input("Nhập số lượng phần tử cho mảng a (n): "))

a = []
print(f"Nhập phần tử cho mảng a:")
for i in range(n):
    while True:
        try:
            element = int(input(f"Nhập phần tử thứ {i+1}: "))
            a.append(element)
            break
        except ValueError:
            print("Giá trị không hợp lệ. Vui lòng nhập một số nguyên.")

b = []
for element in a:
    if element % 2 == 0:
        b.append(element)

print(f"Mảng a: {a}")
print(f"Mảng b (các phần tử chẵn của a): {b}")
```

```python
"""Bai 13 """
def merge_arrays(arr1, arr2):
    merged_array = []
    len1 = len(arr1)
    len2 = len(arr2)
    min_len = min(len1, len2)


    for i in range(min_len):
        merged_array.append(arr1[i] + arr2[i])

    if len1 > len2:
        merged_array.extend(arr1[min_len:])
    elif len2 > len1:
        merged_array.extend(arr2[min_len:])

    return merged_array


a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

result = merge_arrays(a, b)

print(f"Mảng a: {a}")
print(f"Mảng b: {b}")
print(f"Mảng kết quả: {result}")


a2 = [1, 2, 3, 4, 5]
b2 = [10, 20]
result2 = merge_arrays(a2, b2)
print(f"\nMảng a2: {a2}")
print(f"Mảng b2: {b2}")
print(f"Mảng kết quả 2: {result2}")

a3 = [1, 2, 3]
b3 = [4, 5, 6]
result3 = merge_arrays(a3, b3)
print(f"\nMảng a3: {a3}")
print(f"Mảng b3: {b3}")
print(f"Mảng kết quả 3: {result3}")

```
